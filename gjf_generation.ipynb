```python

def right_spacing(coor_list):
    coor_list_space=[]
    for cline in coor_list:
        split_coor=cline.split()
        new_coor=''
        space1=18
        space2=4
        space3=4
        if len(split_coor[0]) == 2:
            space1=17
        for o in split_coor:  
            if '-' in o:
                index_list=[]
                index_list.append(split_coor.index(o))
                if 1 in index_list:
                    space1=17
                elif 2 in index_list:
                    space2=3
                elif 3 in index_list:
                    space3=3
        new_coor=str(split_coor[0])+space1*' '+str(split_coor[1])+'000'+space2*' '+str(split_coor[2])+'000'+space3*' '+str(split_coor[3])+'000\n'
        coor_list_space.append(new_coor)
    return coor_list_space
```

```python

def copy_structure(filename,keywords,space,suffix,numbering,by_num='No'):
    # set by_num to 'yes' if you want the name of gjf to be filename (xyz file) + index of the conformer by order in the xyz file
    # otherwise (follow by else: line below), the gjf will be named after the molecule name in the xyz file + suffix 
    xyz_output=[]

    with open(filename) as output:
        for line in output:
            xyz_output+= [line]

    Eopt_index_list =[i+1 for i,d in enumerate(xyz_output) if len(d)<5]
    
    Eopt_list=[]
    for i in Eopt_index_list:
        Eopt_list.append(xyz_output[i])

    division1=[]
    for j in range(0,len(Eopt_index_list)):
        try:
            a=xyz_output[Eopt_index_list[j]+1: Eopt_index_list[j+1]-1]
            division1.append(a)
        except IndexError:
            b=xyz_output[Eopt_index_list[j]+1:]
            division1.append(b)

    division2=[]
    division3=[]
    for s in division1:
        division2.append(right_spacing(s))

    for structure in division2: 
        coor_content=''
        for individual_structure in structure:
            coor_content+=str(individual_structure)
        content=keywords+coor_content+space
        division3.append(content)

    
    if by_num=='No':
        
        filename=[]
        for indv in Eopt_list:
            filename_list=indv.split()
            filename.append(filename_list[0]+'_'+suffix+'.gjf')
        
            #double check
        for b,a in zip(filename,division3):
            print(b,'\n',a)

        #write the file
        for ifilename, gcontent in zip(filename,division3):
            file = open(ifilename, "w") 
            file.write(gcontent) 
            file.close()

    else:
                    
        select=[y-1 for y in numbering]
        select1=[y+1 for y in select]

        division4=[]
        for o in select:
            division4.append(division3[o])
    
        new_filename= filename[:-4]
    
        filename_list2=[]
        for num2 in select1:
            filename_list2.append(new_filename+'_'+str(num2)+'.gjf')
    
        for c,d in zip(filename_list2,division4):
            print(c,d)
            file = open(c, "w") 
            file.write(d) 
            file.close() 
    
    return 
```

```python
keywords='''%nprocshared=12
%mem=8GB
# opt freq b3lyp/6-31g(d) int=ultrafine

Title Card Required

1 1
''' 
keywords2='''%nprocshared=12
%mem=2GB
# wb97xd/6-311g(d,p) int=ultrafine 

Title Card Required

1 2
''' 

space='''
'''

suffix='spe'
```

```python
numbering=[1,3]
copy_structure('C_Me_RMe_R.xyz',keywords,space,suffix,numbering,by_num='yes')
```
